```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("brain_stroke.csv")
print(df.head())
```

```python
# Get information about the dataframe, including the data types and non-null values
print(df.info())
```

```python
# Get summary statistics for numerical columns
print(df.describe())
```

```python
# Get summary statistics for categorical columns
print(df.describe(include=['object']))
```

```python
# Get the shape of the dataframe (number of rows and columns)
print(df.shape)
```

```python
# Display the names of the columns
print(df.columns)
```

```python
print(df.isnull())  # Check for missing values
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Plot histograms of numerical columns
df.hist(figsize=(10, 10))
plt.show()
```

```python
# Plot correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title('Box Plot of All Numerical Columns')
plt.show()
```

```python
# Scatter Plot (assuming you want to plot 'column_x' vs 'column_y')
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='bmi', data=df)
plt.title('Scatter Plot of column_x vs column_y')
plt.show()
```
