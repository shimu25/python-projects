```python
# Set plot style
sns.set(style="whitegrid")
```

```python
#Distribution of Math, Reading, Writing Scores
plt.figure(figsize=(10, 6))
sns.histplot(df["math score"], kde=True, color="blue", label="Math", bins=20)
sns.histplot(df["reading score"], kde=True, color="green", label="Reading", bins=20)
sns.histplot(df["writing score"], kde=True, color="red", label="Writing", bins=20)
plt.legend()
plt.title("Distribution of Scores")
plt.show()
```

```python
#  Boxplot of scores by gender
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="gender", y="math score")
plt.title("Math Score by Gender")
plt.show()
```

```python
#  Average scores by race/ethnicity
plt.figure(figsize=(10, 6))
df_grouped = df.groupby("race/ethnicity")[['math score', 'reading score', 'writing score']].mean().reset_index()
df_grouped.plot(x="race/ethnicity", kind="bar", figsize=(10, 6))
plt.title("Average Scores by Race/Ethnicity")
plt.ylabel("Average Score")
plt.show()
```

```python
#  Effect of test preparation course
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="test preparation course", y="math score")
plt.title("Math Score vs Test Preparation Course")
plt.show()
```

```python
#  Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df[["math score", "reading score", "writing score"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Scores")
plt.show()
```

```python
#  Scatterplot between math and reading scores
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="math score", y="reading score", hue="gender", style="test preparation course")
plt.title("Math vs Reading Scores")
plt.show()
```

```python

#  Pairplot for all three scores
sns.pairplot(df[["math score", "reading score", "writing score", "gender"]], hue="gender")
plt.suptitle("Pairplot of Scores", y=1.02)
plt.show()
```
