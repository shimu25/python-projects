```python
import numpy as np
# Define the sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Initialize random weights and bias
np.random.seed(0)
weights = np.random.rand(14)
bias = np.random.rand()
# Define the learning rate
learning_rate = 0.001
print(weights)
print(bias)
```

```python

# Define the number of epochs (iterations)
num_epochs = 250

# Training the neural network
for epoch in range(num_epochs):
    # Forward propagation
    z = np.dot(x_train, weights) + bias
    predicted_labels = sigmoid(z)

    # Calculate the error
    error = y_train - predicted_labels

    # Backpropagation
    d_weights = np.dot(x_train.T, error)
    d_bias = np.sum(error)

    # Update the weights and bias
    weights += learning_rate * d_weights
    bias += learning_rate * d_bias
```

```python
# Test the trained model on the same dataset
z = np.dot(x_test, weights) + bias
predicted_labels = sigmoid(z)
predicted_labels = (predicted_labels >= 0.5).astype(int)


# Calculate accuracy
accuracy = np.mean(predicted_labels == y_test)
print("Accuracy:", accuracy)
```
